```python
dir0 = '/home/cloudera/2.kkbox_churn/data01/feature_creation_01/'
infile1 = dir0 + 'members'
infile2 = dir0 + 'transactions'
infile3 = dir0 + 'userlogs.snappy.parquet'

df1 = spark.read.format('csv').option('header','true').load(infile1)
df2 = spark.read.format('csv').option('header','true').load(infile2) \
      .withColumnRenamed('msno', 'msno_')
df3 = spark.read.format('parquet').load(infile3)
```

```python
df3.printSchema()
```

```python
df_j1 = df1.join(df2, df1['msno']==df2['msno_'], 'inner').drop('msno_')
df_j2 = df_j1.join(df3, df_j1['msno']==df3['msno_12w'], 'inner').drop('msno_12w')
```

```python
outfile = dir0 + 'all_features'
df_j2.write.format('parquet').save(outfile)
```
